```python
def reshape_img(fl):
    fl_ = np.zeros((256,256,fl.shape[0]))
    for c in range(fl.shape[0]):
        fl_[:,:,c] = fl[c,:,:]
    return fl_

def clean_br(img):
    img = img - img.min()
    img = img/img.max()
    return img

def clean_fl(img):
    img = reshape_img(img)
    img_ = np.zeros_like(img)
    print(img.shape)
    print(img_.shape)
    for c in range(img.shape[-1]):
        img_[:,:,c] = img[:,:,c] - img[:,:,c].min()
        img_[:,:,c] = img_[:,:,c]/img_[:,:,c].max()
    return img_
```

```python
s=1; z=3; p=0;
br = clean_br(data[0,z,0,p,:,:])
fl = clean_fl(data[0,z,1:,p,:,:])
```

```python
plt.figure(figsize=(15,15))
plt.subplot(1,2,1)
_ = plt.imshow(br, cmap='gray')
plt.subplot(1,2,2)
_ = plt.imshow(fl)
```
